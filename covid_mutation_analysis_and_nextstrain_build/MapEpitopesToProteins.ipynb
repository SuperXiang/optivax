```python
'''Small script to see what proteins each epitope is on'''


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_epi = pd.read_csv('data/processed/epitope_entropies.csv')
df_epi.head()
```

```python
vaccine_epitopes = ['AAISDYDYYRY', 'EEFEPSTQYEY', 'FACPDGVKHVY', 'FPFTIYSLL', 'FTSDYYQLY', 
                   'FVAAIFYLI', 'IFFASFYYVW', 'KVAGFAKFLK', 'LEMELTPVV', 'WSMATYYLF']
```

```python
seqs = df_epi[df_epi.sequence.isin(vaccine_epitopes)][['protein', 'sequence', 'epi_start_pos']]
seqs.head()
```

```python
# want these to be sorted a particular way
new_df = []
for v in vaccine_epitopes:
    new_df.append( [seqs[seqs.sequence == v].iloc[0].protein, 
                    seqs[seqs.sequence == v].iloc[0].sequence, seqs[seqs.sequence == v].iloc[0].epi_start_pos])
    
```

```python
df = pd.DataFrame(new_df, columns=['Protein', 'Epitope', 'Epitope Position'])
# df.head()
df
```
